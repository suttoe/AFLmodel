General Set Up
 1. Check Requirements 

Couple things to check before we get started:

1.Python installed with a version >= 3.5

2. R installed with a version >= 3.2

3.Have Anaconda/Jupyter installed

 
2. Installing R within Jupyter Notebook

    Open the Anaconda Prompt

    Run the following command line:

conda install -c r r-essentials

This will install 100+ R libraries to get us started. From this point we should be able to run a R session in Jupyter by selecting an R kernel. It install an other R libraries, it is easiest to do so through R studio using the install.package('package') method. 

 
Preparing to Use Python and R in the Same Jupyter Note

 
1. Install rpy2

rpy2 is the library that allows use to use R in a python session, this requires a little bit of set up first

    In the anaconda prompt install rpy2 using pip:

pip install rpy2

    In the PATH environment variable, add the path to the R bin

        Find the path to the R bin, might look something like:

C:\Program Files\R\R-3.6.3\bin

b. Navigate to environment variables: 

Control Panel > System > Advanced system settings > Environment Variables…

c. Select Path > Edit… > New > add R bin path > ok

d. Still in the environment variable page, we are going to add some new variables. Click of New.. for to create those listed below:

-R_HOME: Root of R e.g. C:\Program Files\R\R-3.6.3

-R_USER: who will use R (if you want to use R somewhere else who will need to add the location to there as well) e.g. C:\Users\<username>\Anaconda3\Lib\site-packages\rpy2

-R_LIBS_USER: where the libraries are located e.g. C:\Users\<username>\Documents\R\win-library\3.6
2. Using R in notebook

Should be now ready to test out using python and R in the same notebook

    Load R using a magic command (only have to do once per notebook)

%load_ext rpy2.ipython 

If when called it returns a UserWarning regarding the symbol ‘quartz’ you can ignore it. Appears to be a bug.

    Set a cell as an R cell using a magic cell:

%%R 

    Sending a variable (including a dataframe, which will be in tibble form) to use in the R cell can be done when setting up the magic cell:

%%R -i var or %%R -i var1,var2                 

(If passing in multiple variable make sure there is no whitespace between the variables only a comma)

    Sending a variable (including a dataframewhihc will be in pandas form) from R to python can be done when setting up the magic cell:

%%R -o var or %%R -o var1,var2 

(Again if passing multiple variables make sure there are no whitespace between them only commas)

 
Conclusion
You should now have the ability to jump between python and R in the same notebook!
